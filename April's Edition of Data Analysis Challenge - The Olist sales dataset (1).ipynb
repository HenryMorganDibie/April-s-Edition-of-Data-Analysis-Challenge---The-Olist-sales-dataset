# What is the average order value (AOV) on Olist, and how does this vary by product category or payment method?

```python
# read in the datasets
orders = pd.read_csv('olist_orders_dataset.csv')
order_payments = pd.read_csv('olist_order_payments_dataset.csv')
order_items = pd.read_csv('olist_order_items_dataset.csv')
products = pd.read_csv('olist_products_dataset.csv')
product_category_translation = pd.read_csv('product_category_name_translation.csv')

# merge the datasets
merged_orders = pd.merge(orders, order_payments, on='order_id', how='left')
merged_orders_items = pd.merge(merged_orders, order_items, on='order_id', how='left')
merged_orders_items_products = pd.merge(merged_orders_items, products, on='product_id', how='left')
merged_orders_items_products_categories = pd.merge(merged_orders_items_products, product_category_translation, on='product_category_name', how='left')

# calculate total revenue and number of orders
total_revenue = merged_orders['payment_value'].sum()
num_orders = merged_orders['order_id'].nunique()

# calculate AOV
aov = total_revenue / num_orders
print(f"The average order value on Olist is ${aov:.2f}")

# calculate AOV by product category
aov_by_category = merged_orders_items_products_categories.groupby('product_category_name')['payment_value'].sum() / merged_orders_items_products_categories.groupby('product_category_name')['order_id'].nunique()
aov_by_category = aov_by_category.sort_values(ascending=False)
print("Average Order Value by Product Category:")
print(aov_by_category)

# calculate AOV by payment method
aov_by_payment = merged_orders.groupby('payment_type')['payment_value'].sum() / merged_orders.groupby('payment_type')['order_id'].nunique()
aov_by_payment = aov_by_payment.sort_values(ascending=False)
print("Average Order Value by Payment Method:")
print(aov_by_payment)
```

# What is the distribution of seller ratings on Olist, and how does this impact sales performance?

### This will give us an idea of the impact of seller location on sales performance. Sellers in states with higher average sales per seller may have a competitive advantage, as they may be able to charge higher prices or sell more products. On the other hand, sellers in states with lower average sales per seller may need to adjust their prices or marketing strategies to remain competitive.

```python
# Load data
sellers = pd.read_csv('olist_sellers_dataset.csv')
orders = pd.read_csv('olist_orders_dataset.csv')
order_items = pd.read_csv('olist_order_items_dataset.csv')

# Merge order items and orders to get total sales for each order
order_data = order_items.merge(orders[['order_id', 'order_purchase_timestamp']], on='order_id', how='inner')
order_data['total_sales'] = order_data['price'] + order_data['freight_value']

# Merge order data and sellers to get total sales for each seller
seller_data = order_data.merge(sellers[['seller_id', 'seller_state']], on='seller_id', how='inner')
seller_data = seller_data.groupby(['seller_id', 'seller_state']).agg(total_sales=('total_sales', 'sum')).reset_index()

# Group sellers by state and calculate average sales per seller
state_sales = seller_data.groupby('seller_state').agg(avg_sales_per_seller=('total_sales', 'mean'), num_sellers=('seller_id', 'nunique')).reset_index()

# Plot average sales per seller against number of sellers in each state
plt.scatter(state_sales['num_sellers'], state_sales['avg_sales_per_seller'])
plt.xlabel('Number of Sellers')
plt.ylabel('Average Sales per Seller')
plt.title('Relationship between Seller Location and Sales Performance')
plt.show()
```

# How many customers have made repeat purchases on Olist, and what percentage of total sales do they account for?

```python
# Load necessary datasets
orders = pd.read_csv('olist_orders_dataset.csv')
order_items = pd.read_csv('olist_order_items_dataset.csv')

# Merge datasets to get customer and product information for each order
order_data = orders.merge(order_items, on='order_id', how='inner')

# Count number of orders per customer
customer_order_counts = order_data.groupby('customer_id')['order_id'].nunique()

# Count number of customers with multiple orders
repeat_customers = customer_order_counts[customer_order_counts > 1]

# Calculate percentage of total sales made by repeat customers
total_sales = order_data['price'].sum()
repeat_sales = order_data[order_data['customer_id'].isin(repeat_customers.index)]['price'].sum()
repeat_sales_pct = repeat_sales / total_sales * 100

print(f'Total number of customers: {len(customer_order_counts)}')
print(f'Number of repeat customers: {len(repeat_customers)}')
print(f'Percentage of total sales made by repeat customers: {repeat_sales_pct:.2f}%')
```

```python
# Load the necessary datasets
orders = pd.read_csv('olist_orders_dataset.csv')
order_payments = pd.read_csv('olist_order_payments_dataset.csv')

# Merge the datasets
df = pd.merge(orders, order_payments, on='order_id')

# Compute the number of orders per customer
order_counts = df.groupby('customer_id')['order_id'].nunique()

# Compute the number of repeat customers
num_repeat_customers = sum(order_counts > 1)

# Compute the total sales and sales from repeat customers
total_sales = df['payment_value'].sum()
repeat_sales = df[df['customer_id'].isin(order_counts[order_counts > 1].index)]['payment_value'].sum()

# Compute the percentage of total sales from repeat customers
repeat_sales_pct = repeat_sales / total_sales * 100

print(f"Number of customers who have made repeat purchases: {num_repeat_customers}")
print(f"Repeat customers account for {repeat_sales_pct:.2f}% of total sales.")
```

# What are the top-selling products on Olist, and how have their sales trends changed over time?

```python
# Load the necessary datasets
order_items = pd.read_csv('olist_order_items_dataset.csv')
orders = pd.read_csv('olist_orders_dataset.csv')
products = pd.read_csv('olist_products_dataset.csv')

# Merge the datasets together
merged_data = order_items.merge(orders[['order_id', 'order_purchase_timestamp']], on='order_id', how='left')
merged_data = merged_data.merge(products[['product_id', 'product_category_name']], on='product_id', how='left')

# Convert the date column to a pandas datetime object
merged_data['order_purchase_timestamp'] = pd.to_datetime(merged_data['order_purchase_timestamp'])

# Get the top-selling products
top_products = merged_data['product_category_name'].value_counts().head(10)

# Create a dictionary to store the sales trends for each top-selling product
sales_trends = {}

# Loop through each top-selling product and get its sales trends
fig, ax = plt.subplots(figsize=(15, 10))
for product in top_products.index:
    # Filter the merged_data for this product
    product_data = merged_data[merged_data['product_category_name'] == product]
    
    # Set the date column as the index
    product_data.set_index('order_purchase_timestamp', inplace=True)
    
    # Resample the data by month and sum the sales
    sales = product_data['price'].resample('M').sum()
    
    # Add the sales trends to the dictionary
    sales_trends[product] = sales

# Plot the sales trends for each top-selling product

for product, sales in sales_trends.items():
    plt.plot(sales.index, sales.values, label=product)

plt.legend()
plt.title('Sales Trends for Top-Selling Products')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.show()
```

# Which product categories have the highest profit margins on Olist, and how can the company increase profitability across different categories?

```python
# Load the necessary datasets
order_items = pd.read_csv('olist_order_items_dataset.csv')
products = pd.read_csv('olist_products_dataset.csv')

# Merge the order items with the products dataset to get the product category and product cost
product_costs = pd.merge(order_items[['order_id', 'product_id', 'price', 'freight_value']], 
                         products[['product_id', 'product_category_name']], on='product_id')

# Calculate the cost of goods sold (COGS) for each order item
product_costs['cogs'] = product_costs['price'] + product_costs['freight_value']

# Group the order items by product category and calculate the total revenue and COGS
category_profit = product_costs.groupby('product_category_name').agg({'cogs': 'sum', 'price': 'sum'})

# Calculate the profit margin for each product category
category_profit['profit_margin'] = (category_profit['price'] - category_profit['cogs']) / category_profit['price']

# Sort the product categories by profit margin and get the top 10
top_categories = category_profit.sort_values(by='profit_margin', ascending=False).head(10)

# Print the top categories and their profit margins
print(top_categories)
```

```python
# Join order_items and products dataframes
merged_order_items = pd.merge(order_items, products, on='product_id', how='inner')

# Calculate the revenue for each order
revenue = pd.merge(merged_order_items, order_payments, on='order_id', how='inner').groupby('order_id')['payment_value'].sum()

# Join the translated product category names
category_revenue = pd.merge(merged_order_items, product_category_name_translation, on='product_category_name', how='inner')
category_revenue = category_revenue.groupby('product_category_name_english')['price', 'freight_value'].sum()

# Calculate profit margin for each category
category_revenue['profit'] = category_revenue['price'] - category_revenue['freight_value'] - revenue.mean()
category_revenue['profit_margin'] = category_revenue['profit'] / category_revenue['price'] * 100

# Sort by profit margin
category_revenue = category_revenue.sort_values(by='profit_margin', ascending=False)

# Plot the results
plt.figure(figsize=(18, 6))
sns.barplot(x=category_revenue.index, y=category_revenue['profit_margin'])
plt.xticks(rotation=90)
plt.xlabel('Product Category')
plt.ylabel('Profit Margin (%)')
plt.title('Profit Margins for Olist Product Categories')
plt.show()
```

# How does Olist's marketing spend and channel mix impact sales and customer acquisition costs, and how can the company optimize its marketing strategy to increase ROI?

```python
# Merge relevant datasets
orders_data = orders.merge(order_payments, on='order_id', how='inner')
orders_data = orders_data.merge(order_items, on='order_id', how='inner')
orders_data = orders_data.merge(order_reviews, on='order_id', how='inner')
orders_data = orders_data.merge(customers, on='customer_id', how='inner')

# Calculate marketing spend for each channel
marketing_data = orders_data.groupby('payment_type')['payment_value'].sum().reset_index()
marketing_data.rename(columns={'payment_type': 'marketing_channel', 'payment_value': 'marketing_spend'}, inplace=True)

# Calculate sales and customer acquisition costs for each channel
sales_data = orders_data.groupby('payment_type')['price'].sum().reset_index()
sales_data.rename(columns={'payment_type': 'marketing_channel', 'price': 'total_sales'}, inplace=True)

customers_data = orders_data.drop_duplicates(subset=['customer_id'])
cac_data = customers_data.groupby('payment_type')['customer_id'].count().reset_index()
cac_data.rename(columns={'payment_type': 'marketing_channel', 'customer_id': 'total_customers'}, inplace=True)

cac_data['cac'] = marketing_data['marketing_spend'] / cac_data['total_customers']
sales_data['cpc'] = marketing_data['marketing_spend'] / sales_data['total_sales']

# Analyze the relationship between marketing spend and sales, as well as the customer acquisition costs for each channel
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

sns.barplot(x='marketing_channel', y='marketing_spend', data=marketing_data, ax=axs[0])
axs[0].set_title('Marketing Spend by Channel')

sns.barplot(x='marketing_channel', y='total_sales', data=sales_data, ax=axs[1])
axs[1].set_title('Total Sales by Channel')

fig, ax = plt.subplots(1, 1, figsize=(6, 5))

sns.barplot(x='marketing_channel', y='cac', data=cac_data, ax=ax)
ax.set_title('Customer Acquisition Cost by Channel')

# Optimize the marketing strategy by allocating resources to the channels that yield the highest ROI
sales_data['roi'] = (sales_data['total_sales'] / marketing_data['marketing_spend']) * 100
max_roi = sales_data.loc[sales_data['roi'].idxmax(), 'marketing_channel']
print('The marketing channel with the highest ROI is', max_roi)
```

# Geolocation having high customer density. Calculate customer retention rate according to geolocations.

```python
# Load the datasets
orders = pd.read_csv('olist_orders_dataset.csv')
order_items = pd.read_csv('olist_order_items_dataset.csv')
customers = pd.read_csv('olist_customers_dataset.csv')
geolocation = pd.read_csv('olist_geolocation_dataset.csv')

# Merge the orders and customers datasets
order_cust = pd.merge(orders, customers, on='customer_id', how='inner')

# Merge the order_cust and geolocation datasets
order_cust_geo = pd.merge(order_cust, geolocation, left_on='customer_zip_code_prefix', right_on='geolocation_zip_code_prefix', how='left')

# Drop unnecessary columns
order_cust_geo = order_cust_geo.drop(['customer_zip_code_prefix', 'geolocation_zip_code_prefix'], axis=1)
```

```python
# Group by geolocation to count the number of customers in each area
cust_by_geo = order_cust_geo.groupby(['geolocation_lat', 'geolocation_lng']).agg({'customer_id': pd.Series.nunique}).reset_index()

# Sort by number of customers in descending order
cust_by_geo = cust_by_geo.sort_values('customer_id', ascending=False)

# Print the top 10 geolocations with the highest customer density
print(cust_by_geo.head(10))
```

```python
# Get the customer data for the top geolocations with high customer density
top_geo = cust_by_geo.head(10)
top_geo_cust = order_cust_geo[order_cust_geo[['geolocation_lat', 'geolocation_lng']].apply(tuple, axis=1).isin(top_geo[['geolocation_lat', 'geolocation_lng']].apply(tuple, axis=1))]

# Identify customers who have made repeat purchases
repeat_cust = top_geo_cust.groupby('customer_id').filter(lambda x: len(x) > 1)['customer_id'].unique()

# Calculate the retention rate of those repeat customers
retention_rate = len(repeat_cust) / len(top_geo_cust['customer_id'].unique()) * 100

print('Retention rate for customers in high-density geolocations: {:.2f}%'.format(retention_rate))
```
